# Inference with Tacotron2

Move to tacotron folder, and import the necessary libraries

Load the tacotron checkpoint & model

Perform an inference using the loaded model

```
%cd Tacotron/
import numpy as np
from IPython.display import Audio
from scipy.io import wavfile
from synthesizer import Synthesizer
```

```
checkpoint_path = '/tmp/tacotron-20180906/model.ckpt'
tacotron2 = Synthesizer()
tacotron2.load(checkpoint_path)
```

```
framerate = 44100
text = "After reading this you should feel comfortable using Python for basic text-to-speech applications"
audio_result = tacotron2.synthesize(text)
Audio(audio_result,rate=framerate)
```
